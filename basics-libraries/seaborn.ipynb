```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
x = np.linspace(0, 14, 100)
```

```python
for i in range(1, 5):
    plt.plot(x, np.sin(x + 0.5*i)*(7-i))
```

```python
sb.set()
for i in range(1, 5):
    plt.plot(x, np.sin(x + 0.5*i)*(7-i))
```

```python
sb.set_style("white")
sb.set_palette("Greens")
for i in range(1, 5):
    plt.plot(x, np.sin(x + 0.5*i)*(7-i))
```

```python
# Distribution Plot
sb.distplot(df['SepalLengthCm'])
```

```python
# Pair Plot
sb.set_style('ticks')
sb.pairplot(df, hue='Species', palette='husl', diag_kind='kde', kind="scatter")
```

```python
# Strip Plot
sb.set_palette('husl')

#jitter for straight line output
sb.stripplot(x='Species', y='PetalLengthCm', data=df)
```

```python
# Swarm Plot
sb.set_palette('husl')
sb.swarmplot(x='Species', y='PetalLengthCm', data=df)
```

```python
sb.boxplot(x='Species', y='PetalLengthCm', data=df)
```

```python
sb.boxplot(data=df, orient='h')
```
