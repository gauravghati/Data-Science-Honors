```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
df = pd.read_csv('../data/kaggle_titanic_dataset.csv', index_col=0)
df.head()
```

   - survival - Survival (0 = No; 1 = Yes)
   - class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)

   - name - Name | sex - Sex | age - Age

   - sibsp - Number of Siblings/Spouses Aboard

   - parch - Number of Parents/Children Aboard
   - ticket - Ticket Number | fare - Passenger Fare | cabin - Cabin
   - embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)



**Data Pre Processing**

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
df.columns
```

```python
df.drop(columns=['Name', 'Cabin', 'Embarked', 'Ticket'], inplace=True)
df.head()
```

```python
df['Sex'] = df['Sex'].apply(lambda x:x=='male')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(value=df['Age'].interpolate(), inplace=True)
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sb.heatmap(df.corr(), annot=True)
```

```python
plt.figure(figsize=(5, 5))
plt.pie(df['Survived'].value_counts(), labels=['Not Survived', 'Survived'], 
        autopct='%.2f', colors=['orange', 'yellow'])

print(df['Survived'].value_counts())
```

```python
sb.pairplot(df, hue='Survived')
```

**Construct and fit classifiction problem**

```python
x = df[['Sex', 'Pclass', 'Age']]
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
test_set_prediction = model.predict(x_test)
test_set_prediction
```

```python
confusion_metrix = metrics.confusion_matrix(y_test, test_set_prediction)
matrix_df = pd.DataFrame(data=confusion_metrix, index=['No', 'Yes'], columns=['Predicted No', 'Predicted Yes'])
matrix_df
```

```python
sb.heatmap(data=matrix_df, cmap='Blues', annot= True, fmt='d')
```

```python
print(metrics.classification_report(y_test, test_set_prediction))
```
