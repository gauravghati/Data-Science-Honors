```python
# For SepalLengthCm
fig, ax = plt.subplots(figsize=(10, 6))
sb.histplot(data=df,x='SepalLengthCm', kde=True, hue='Species', bins=15)
```

```python
# SepalWidthCm
fig, ax = plt.subplots(figsize=(10, 6))
sb.histplot(data=df,x='SepalWidthCm', kde=True, hue='Species', bins=15, ax=ax)
```

```python
# PetalLengthCm
fig, ax = plt.subplots(figsize=(10, 6))
sb.histplot(data=df,x='PetalLengthCm', kde=True, hue='Species', bins=20, ax=ax)
```

```python
# PetalWidthCm
fig, ax = plt.subplots(figsize=(10, 6))
sb.histplot(data=df,x='PetalWidthCm', kde=True, hue='Species', bins=20, ax=ax)
```

```python
fig,a = plt.subplots(2,2,figsize=(14,16))
sb.boxplot(data=df,y='SepalLengthCm', x='Species',ax=a[0][0])
sb.boxplot(data=df,y='SepalWidthCm', x='Species',ax=a[0][1])
sb.boxplot(data=df,y='PetalLengthCm', x='Species',ax=a[1][0])
sb.boxplot(data=df,y='PetalWidthCm', x='Species',ax=a[1][1])
```
